**Matplotlib is a 2D and 3D visualization / plotting library. Creates high quality graphs (histograms, bar charts, scatterplots, areaplots, pyplots) to display information.**

```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Lineplot**

```python
x = np.arange(0, 10)
x
```

```python
y = 2 * x
y
```

```python
z = 3 * x
z
```

```python
plt.plot(x, y, color = "r", marker = "*", linestyle = "--", linewidth = 3, markersize = 10)
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("title")
plt.axis([0, 10, 0, 20])
plt.legend(["line one"], loc = 1)
plt.show()
```

```python
plt.plot(x, y, "bo--", linewidth = 3, markersize = 10, label = "x by y", markerfacecolor = "r", markeredgewidth = 1, markeredgecolor = "black")
plt.plot(x, z, "r*-", linewidth = 3, markersize = 10, label = "x by z", markerfacecolor = "b", markeredgewidth = 2, markeredgecolor = "black")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("title")
plt.axis([0, 10, 0, 20])
plt.legend(loc = 1)
plt.savefig("graph.png")
plt.grid(color = "black",linestyle = "--", linewidth = 2)
plt.show()
```

```python
fig = plt.figure()
axis = fig.add_axes([0, 0, 1, 1])

# left_bottom, width, height

axis.plot(x, y)
plt.show()
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0, 0, 1, 1])
axis2 = fig.add_axes([0.2, 0.2, 0.5, 0.5])

# left_bottom, width, height

axis1.plot(x, y)
axis2.plot(x, y)

axis1.set_xlabel("Xlabel")
axis1.set_ylabel("Ylabel")
axis1.set_title("Title")

axis2.set_xlabel("xlabel")
axis2.set_ylabel("ylabel")
axis2.set_title("title")

plt.show()
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0, 0, 1, 1])
axis2 = fig.add_axes([0.2, 0.2, 0.5, 0.5])

# left_bottom, width, height

axis1.plot(x, y)
axis2.plot(x, y)

axis1.set_xlabel("Xlabel")
axis1.set_ylabel("Ylabel")
axis1.set_title("Title")

axis2.set_xlabel("xlabel")
axis2.set_ylabel("ylabel")
axis2.set_title("title")

axis2.set_xlim(8, 10)
axis2.set_ylim(12, 15)

plt.show()
plt.savefig("plot.png")
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0, 0, 1, 1])
axis2 = fig.add_axes([1, 1, 0.5, 0.5])

# left_bottom, width, height

axis1.plot(x, y)
axis2.plot(x, y)

axis1.set_xlabel("Xlabel")
axis1.set_ylabel("Ylabel")
axis1.set_title("Title")

axis2.set_xlabel("xlabel")
axis2.set_ylabel("ylabel")
axis2.set_title("title")

axis2.set_xlim(8, 10)
axis2.set_ylim(1000, 1500)

plt.show()
plt.savefig("plot.png")
```

```python
fig, axis = plt.subplots(nrows = 2, ncols = 2)

axis[0][0].plot(x, y)
axis[0][1].plot(x, z)

axis[1][0].plot(y, z)
axis[1][1].plot(z, x)

axis[0][0].set_xlabel("0, 0, Xlabel")
axis[0][0].set_ylabel("0, 0, Ylabel")
axis[0][0].set_title("0, 0, Title")

axis[0][1].set_xlabel("0, 1, Xlabel")
axis[0][1].set_ylabel("0, 1, Ylabel")
axis[0][1].set_title("0, 1, Title")

axis[1][0].set_xlabel("1, 0, Xlabel")
axis[1][0].set_ylabel("1, 0, Ylabel")
axis[1][0].set_title("1, 0, Title")

axis[1][1].set_xlabel("1, 1, Xlabel")
axis[1][1].set_ylabel("1, 1, Ylabel")
axis[1][1].set_title("1, 1, Title")

fig.subplots_adjust(left = 0.5, bottom = 0.5, wspace = 1.5, hspace = 3)
```

***MatPlotLib - Histograms***

```python
mlStudentsAge = np.random.randint(14, 45, 100)
mlStudentsAge
```

```python
pythonStudentsAge = np.random.randint(10, 45, 100)
pythonStudentsAge
```

```python
plt.hist(mlStudentsAge)

plt.xlabel("Students Age Category")
plt.ylabel("Number of Students")
plt.title("ML Students Age Histogram")

plt.show()
```

```python
Agebins = [15, 20, 25, 30, 35, 40, 45]

plt.hist(mlStudentsAge, bins = Agebins, rwidth = 0.5, color = "Black", label = "ml student")
plt.legend()

plt.xlabel("Students Age Category")
plt.ylabel("Number of Students")
plt.title("ML Students Age Histogram")

plt.show()
```

```python
plt.hist([mlStudentsAge, pythonStudentsAge], bins = Agebins, rwidth = 0.5, color = ["Black", "Red"], label = ["ML student", "Python student"])
plt.legend()

plt.xlabel("Students Age Category")
plt.ylabel("Number of Students")
plt.title("ML / Python Students Age Histogram")

plt.show()
```

```python
classes = ["Python", "AI", "ML", "Java", "C++"]

class1 = [40, 10, 52, 32, 45]

plt.bar(classes, class1)
plt.show()
```

```python
classes = ["Python", "AI", "ML", "Java", "C++"]

class1 = [40, 10, 52, 32, 45]

plt.barh(classes, class1)
plt.show()
```

```python
classes = ["Python", "AI", "ML", "Java", "C++"]

class1 = [40, 10, 52, 32, 45]

plt.bar(classes, class1, width = 0.2, align = "edge", color = "Black", edgecolor = "Red", alpha = 0.25, label = "Class Students", linestyle = "--")
plt.legend()
plt.show()
```

```python
classes = ["Python", "AI", "ML", "Java", "C++"]

class1 = [69, 54, 52, 32, 45]
class2 = [32, 34, 43, 42, 19]

class_index = np.arange(len(classes))
width = 0.2

plt.bar(class_index, class1, width, align = "edge", color = "Black", edgecolor = "Red", alpha = 0.25, label = "Class1 Students", linestyle = "--")
plt.bar(class_index + width, class2, width, align = "edge", color = "Red", edgecolor = "Black", alpha = 0.75, label = "Class2 Students", linestyle = "-")

plt.xticks(class_index + width, classes, rotation = 20)

plt.xlabel("Professions")
plt.ylabel("Students")
plt.title("Students in various Professions")

plt.legend()
plt.show()
```

**Pie Plot**

```python
classes = ["Python", "AI", "ML", "Java", "C++"]

class1 = [69, 54, 52, 32, 45]

explode = [0, 0, 0.2, 0, 0]
colors = ["Orange", "Blue", "Red", "Green", "Yellow"]

textProps = {"fontsize": 12}

plt.pie(class1, labels = classes, explode = explode, colors = colors, textprops = textProps, autopct = "%0.2f%%", shadow = True, radius = 2, startangle = 90, rotatelabels = True)

plt.show()
```
