```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = np.arange(0, 10)
c = 3 * 10**8
E = m * c**2
```

```python
plt.plot(m, E, label='E=mc^2')
plt.xlabel('Mass (grams)')
plt.ylabel('Energy (Joules)')
plt.title('Energy-Mass Equivalence')
plt.legend()
plt.show()
```
